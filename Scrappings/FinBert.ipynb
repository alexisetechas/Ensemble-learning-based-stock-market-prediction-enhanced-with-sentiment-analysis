```python
import pandas as pd
import torch
from transformers import pipeline
from tqdm.auto import tqdm
import numpy as np
```

```python
device = 0 if torch.cuda.is_available() else -1
pipe = pipeline(
    "sentiment-analysis", 
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student", 
    device=device,
    truncation=True, 
    max_length=512
)

df = pd.read_csv('moneyTimesFormatado.csv')

df['Titulo'] = df['Titulo'].fillna("").astype(str)
print(f"Total de linhas para processar: {len(df)}")
textos = df['Titulo'].tolist()

score = []
BATCH_SIZE = 64

print("Iniciando an치lise de sentimento em massa...")

for out in tqdm(pipe(textos, batch_size=BATCH_SIZE), total=len(textos)):
    label = out['label']
    score = out['score']
    
    var = 0
    if label == 'positive':
        var = score
    elif label == 'negative':
        var = -score
    # Se for neutral, continua 0
    
    score.append(var)

df['Sentiment_Score'] = score
df.to_csv('FinBert.csv', index=False)
print(df[['Titulo', 'Sentiment_Score']].head(10))
```
